1. Use matplotlib to plot the following equation:

y = x**2 âˆ’ x + 2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
# Calculate y values based on the equation
x = np.linspace(-10, 10, 100)
y = x**2 - x + 2




# Create the plot
plt.plot(x, y, color= 'pink', ls='--') 

plt.xlabel('x')
plt.ylabel('y')
plt.title('Matplot of y = x^2 - x + 2')
plt.grid(True)

# Add annotation for the point (0, 0)
plt.annotate('(0, 0)', xy=(0, 0), xytext=(3, 30))
             #textcoords='offset points', arrowprops=dict(arrowstyle='->'))

# Show the plot
plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# Set up the figure with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2)

# Plot: y = sqrt(x)
x = np.linspace(0, 10, 100)
y1 = np.sqrt(x)
axs[0, 0].plot(x, y1, color='gray', label='plot 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].set_title('Plot of y = sqrt(x)')

# Plot: y = x^3
y2 = x**3
axs[0, 1].plot(x, y2, color='c', label='plot 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].set_title('Plot of y = x^3')

# Plot: y = 1/(x + 1)
y4 = 1 / (x + 1)
axs[1, 0].plot(x, y4, color='y', label='plot 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].set_title('Plot of y = 1/(x + 1)')

# Plot: y = 2x
y3 = 2 * x
axs[1, 1].plot(x, y3, color='m',label='plot 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].set_title('Plot of y = 2x')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
fig.legend(loc='outside upper right')
plt.suptitle('THE PLOTS')
fig.tight_layout()
plt.show()
```
